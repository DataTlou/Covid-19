```python
SA = country[country['Country/Region'] == 'South Africa']
SA
```

```python
SA['Month'] = SA['Date'].apply(lambda x: x.split('/')[0])
SA.head()
```

```python
SA['Month'] = SA['Month'].astype(int)
SA
```

```python
def season(mon):
    if mon >= 3 and 5 >= mon:
        return 'Autumn'
    elif mon >= 6 and mon <= 8:
        return 'Winter'
    elif mon >= 9 and mon <= 11:
        return 'Spring'
    elif mon == 12 or mon <= 2:
        return 'Summer'
```

```python
SA['Season'] = SA['Month'].apply(lambda x: season(x))
```

```python
SA
```

```python
SA[SA['Season'] == 'Summer']
```

```python
def lockdown(x):
    if x > '07/26/2021':
        return 'Level 3'
    elif x <= '07/25/2021' and x >= '06/28/2021':
        return 'Level 4'
    elif x <= '06/27/2021' and x >= '06/16/2021':
        return 'Level 3'
    elif x <= '06/15/2021' and x >= '05/31/2021':
        return 'Level 2'
    elif x <= '05/30/2021' and x >= '03/01/2021':
        return 'Level 1'
    elif x <= '02/28/2021' and x >= '12/29/2020':
        return 'Level 3'
    elif x <= '12/28/2020' and x >= '09/21/2020':
        return 'Level 1'
    elif x <= '09/20/2020' and x >= '08/18/2020':
        return 'Level 2'
    elif x <= '08/17/2020' and x >= '06/01/2020':
        return 'Level 3'
    elif x <= '05/31/2020' and x >= '05/01/2020':
        return 'Level 4'
    elif x <= '04/30/2020' and x >= '03/26/2020':
        return 'Level 5'
    else:
        return 'Level 1'
```

```python
SA['Lockdown'] = SA['Date'].apply(lambda x: lockdown(x))
SA.tail()
```

```python
def alcohol(x):
    if x >= '03/27/2020' and x <= '06/01/2020':
        return 'Banned'
    elif x >= '07/12/2020' and x <= '08/17/2020':
        return 'Banned'
    elif x >= '12/15/2020' and x <= '02/01/2021':
        return 'Banned'
    elif x >= '06/28/2021'and x <= '07/25/2021':
        return 'Banned'
    else:
        return 'Allowed'
```

```python
SA['Alcohol Sales'] = SA['Date'].apply(lambda x: alcohol(x))
SA.head()
```
